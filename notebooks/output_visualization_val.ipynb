```python
import os
import glob
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import random
```

```python
with open('/data/val.txt', 'r') as f:
    filenames = [x.strip() for x in f.readlines()]
```

```python
# Visualize images
data_dir = "/data/train_images/"
anno_dir = "../output"

idx = random.randint(0, len(filenames))
img = np.array(Image.open(os.path.join(data_dir, "%s.tiff" % filenames[idx])).resize([1024, 1024]))
mask = np.array(Image.open(os.path.join(anno_dir, "%s_output.png" % filenames[idx])))

print(f'filename: {filenames[idx]}')
print(f'image min: {np.min(img)}, image max: {np.max(img)}')
print(f'image shape: {img.shape} dtype: {img.dtype}')
print(f'mask min: {np.min(mask)}, mask max: {np.max(mask)}')
print(f'mask shape: {mask.shape} dtype: {img.dtype}')

plt.imshow(img)
plt.imshow(mask, alpha=0.2)
plt.show()
```
